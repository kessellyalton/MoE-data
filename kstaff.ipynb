# Table 4: Number of Teachers by Level , District and Sex

```python
kstaff_c = kstaff.iloc[:,[37]]
kstaff_q = kstaff.iloc[:,[23]]
kstaff_s = kstaff.iloc[:,[36]]
kstaff_o = kstaff.iloc[:,[54]]
kstaff_t = kstaff.iloc[:,15:19]
kstaff_g = kstaff.iloc[:,[2]]
kstaff_o = kstaff.iloc[:,[54]]
kstaff_d = kstaff.iloc[:,38:54]
kstaff_d.head()
```

```python
kece_cds = pd.concat([kstaff_s,kstaff_c,kstaff_o,kstaff_q, kstaff_t,kstaff_g, kstaff_d],axis=1)
kece_cds 


#.groupby(['County','Type of school', 'Professional qualification']).sum().sum(1)#.reset_index()
#teach_qc1 = teach_qc.reset_index()
#teach_qc1[teach_qc1['Type of school']== 'Public']
```

```python
kece_cds.columns
```

```python
kece_melt =kece_cds.melt(id_vars=['School Name', 'County', 'Type of school', 'Professional qualification',
       'Level(s) teaching/ECE', 'Level(s) teaching/Primary (Lower Basic)',
       'Level(s) teaching/Junior Secondary (Upper Basic)',
       'Level(s) teaching/Senior High', 'Sex'],var_name = 'district',value_name='count')
kstaff_tdg = kece_melt.groupby(['County','Sex', 'count']).sum().reset_index()
kstaff_tdg.head()
```

```python
kstaff_tdg.iloc[:,[0,1,2,6]][kstaff_tdg['Sex']== 'Female']
```

# Table 5: Number of Teachers by District and Qualifications

```python
kece_cd = pd.concat([kstaff_s,kstaff_c,kstaff_t,kstaff_q, kstaff_d],axis=1)
kece_cd.columns 
```

```python
kece_melto =kece_cd.melt(id_vars=['School Name', 'County', 'Level(s) teaching/ECE',
       'Level(s) teaching/Primary (Lower Basic)',
       'Level(s) teaching/Junior Secondary (Upper Basic)',
       'Level(s) teaching/Senior High', 'Professional qualification'],var_name = 'district',value_name='count')
kstaff_tdl = kece_melto.groupby(['County','count','Professional qualification']).sum()#.reset_index()
kstaff_tdl.head()
```
